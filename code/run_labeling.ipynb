# Step 2 — run the labeling GUI

Run **after** `attach_data.ipynb` (which attached the data and wrote
`/scratch/label_candidates.csv`). This launches the `roi-classifier-label` window **on the
workstation desktop** over the candidate list.

- This cell **blocks while the GUI is open** — that's expected. Label with the keys below,
  press **`q`** to quit, then the cell finishes.
- Labels are saved to **`/scratch/labels`** (persistent) — *not* `/results` (ephemeral here).
- Keys: **`g`/`b`/`o`/`e`/`u`** = good / bad / bad_ok / merged / unsure · **`s`** skip ·
  **`z`** undo · **`n`/`p`** next/prev subject · **`1`/`2`/`3`** channel · **`m`** MIP ·
  arrows / mouse-wheel scroll z.

**+100 more:** re-run cell 3 of `attach_data.ipynb` (excludes what you just labeled), then
re-run this cell. When fully done, run `create_label_asset.ipynb` to publish.

_(Under the hood this runs `code/label.sh`, which rebuilds the tight-bbox cache then opens
the GUI. Equivalent terminal command: `bash code/label.sh "<sids>"`.)_

```python
import os, subprocess
from pathlib import Path

SUBJECT_IDS = ['790322']      # <-- same subject(s) you attached
REVIEWER    = 'Jinho'     # <-- your name (recorded in every label)

cand = Path('/root/capsule/scratch/label_candidates.csv')
assert cand.exists(), 'No candidate list — run attach_data.ipynb (cells 1-3) first.'

env = {**os.environ, 'REVIEWER': REVIEWER}
print('Launching GUI on the workstation desktop — label, then press q to quit...\n', flush=True)
rc = subprocess.run(['bash', '/root/capsule/code/label.sh', ','.join(SUBJECT_IDS)], env=env)
print(f'\nGUI exited (code {rc.returncode}). Labels written under /scratch/labels.')
print('Add +100 more? re-run attach_data.ipynb cell 3, then re-run this cell.')
print('Done labeling? run create_label_asset.ipynb to publish.')
```
